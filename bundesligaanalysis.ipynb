```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
soccer = pd.read_csv("https://raw.githubusercontent.com/Crash105/Bundesliga/refs/heads/main/bundesliga_player.csv")
soccer
```

```python
map = {

}
count = 0
countNan = 0
for lab, row in soccer.iterrows():
    count += 1
    if pd.isna(row['outfitter']):
        countNan += 1
        continue
    if row['outfitter'] in map:
       map[row['outfitter']]+=1
    else:
        map[row['outfitter']] = 1
print(map)
total_sponsors = count - countNan
print(total_sponsors)
bar = [countNan,count-countNan]
plt.pie(bar, labels = ['Not Sponsored', 'Sponsored'],autopct='%1.1f%%' )
plt.title('Not Sponsored vs Sponsorship')
plt.show()

```

```python
filter = soccer['outfitter'] == 'adidas'
soccer[filter]
adidas_count = 0
for lab, row in soccer[filter].iterrows():
    adidas_count += 1
print(adidas_count)

```

```python
filter = soccer['outfitter'] == 'Puma'
soccer[filter]
puma_count = 0
for lab, row in soccer[filter].iterrows():
    puma_count += 1
print(puma_count)
```

```python
filter = soccer['outfitter'] == 'Nike'
soccer[filter]
nike_count = 0
for lab, row in soccer[filter].iterrows():
    nike_count += 1
print(nike_count)
```

```python
bar = [nike_count,adidas_count,puma_count]
plt.pie(bar,labels=['Nike','Adidas','Puma'],autopct='%1.1f%%')
plt.title('Highest Sponsorship marketshare among top 3 sponsors in German Bundesliga')
plt.show()
```
