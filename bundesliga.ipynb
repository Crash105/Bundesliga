```python
import pandas as pd
```

```python
soccer = pd.read_csv("https://raw.githubusercontent.com/Crash105/Bundesliga/refs/heads/main/bundesliga_player.csv")
```

```python
soccer
```

```python
soccer = soccer.dropna(subset = ["outfitter"])
```

```python
soccer
```

```python
soccer["outfitter"].value_counts()
```

```python
# prompt: how to turn non adidas nike and puma rows to simply other
soccer.loc[~soccer['outfitter'].isin(['adidas', 'Nike', 'Puma']), 'outfitter'] = 'other'
```

```python
soccer["outfitter"].value_counts()
```

```python
soccer
```

```python
soccer.head(100)
```

```python
import matplotlib.pyplot as plt

outfitter_counts = soccer["outfitter"].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(outfitter_counts, labels=outfitter_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Outfitter Distribution")
plt.show()
```
